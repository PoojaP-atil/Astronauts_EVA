```python
df['Crew']=df.Crew.str.replace('Doug Wheellock Tracy Caldwell Dyson',
                               'Doug Wheellock Tracy Caldwell Dyson')
```

```python
df['Crew']=df.Crew.replace("\s\s",",",regex=True)
df.Crew.value_counts()
```

```python
df['Astronaut']=df['Crew'].str.split(',')
astronauts={crew for crew in df['Astronaut'].str[0]}
second={crew for crew in df['Astronaut'].str[1]}
astronauts=astronauts.union(second)
```

```python
astronauts=pd.DataFrame(astronauts,columns=['Crew'])
astronauts['EVA_number']=0
astronauts['EVA_time']=0
astronauts.head()
```

```python
astronauts.dropna(how='any',inplace=True)
astronauts.head()
```

```python
df.head()
```

```python
for Crew in astronauts.Crew:
  for index, row in df.iterrows():
    for astro in row['Astronaut']:
      if astro == Crew:
        astronauts['EVA_number'].loc[astronauts.Crew==Crew]+=1
        astronauts['EVA_time'].loc[astronauts.Crew==Crew]+=row.Duration
astronauts.head()
```

```python
def graph_eva(column):
  astronauts.sort_values(by=column,ascending=False,inplace=True)
  plt.figure(figsize=(18,48))
  sns.barplot(data=astronauts,x=column,y='Crew',palette='crest_r')
  plt.ylabel(None)
  plt.show()
```

```python
graph_eva('EVA_number')
```

```python
graph_eva('EVA_time')
```
